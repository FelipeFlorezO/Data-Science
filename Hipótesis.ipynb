### Importar las librerías necesarias para el análisis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as st
from scipy.stats import f_oneway
from scipy.stats import ttest_ind
```

### De nuevo, usaremos la base de datos iris

```python
iris = sns.load_dataset("iris")
iris.head()
```

### Tipos de pruebas:

### a) Distribucion de t student

```python
se_length = iris.std()['sepal_length']/np.sqrt(10)
se_width = iris.std()['sepal_width']/np.sqrt(10)
```

```python
sed = np.sqrt((se_length**2)+(se_width**2))
```

### Ahora el paso final:

```python
t_stat = (iris.mean()['sepal_length']-iris.mean()['sepal_width'])/sed
t_stat
```

```python
p = ttest_ind(iris['sepal_length'], iris['sepal_width'])
p
```

### b) Correlación y coeficiente de Pearson

```python
iris['sepal_length'].corr(iris['sepal_width'])
```

```python
st.pearsonr(iris['sepal_length'], iris['sepal_width'])
```

### c) ANOVA

```python
f_oneway(iris['sepal_length'], iris['sepal_width'])
```

### Bootstrapping

```python
import random
```

```python
data = np.random.normal(loc = 34, size = 10000)
data
```

```python
data.mean()
```

### Iniciamos bootstrapping para calcular el promedio de edades

```python
promedio = []
for i in range(40):
  muestra = random.sample(data.tolist(),5)
  prom = np.mean(muestra)
  promedio.append(prom)
```

```python
np.mean(promedio)
```

```python
x = df.iloc[:, 2:]
y = df.iloc[:,1]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=4)
modelo = DecisionTreeClassifier()
modelo.fit(x_train, y_train)
resultado = modelo.score(x_test, y_test)
print(resultado)
```

```python
from sklearn.model_selection import KFold

modelo = DecisionTreeClassifier()
kfold_validacion=KFold(10)
```

```python
from sklearn.model_selection import cross_val_score

resultados = cross_val_score(modelo, x, y, cv=kfold_validacion)
print(resultados)
resultados.mean()
```
